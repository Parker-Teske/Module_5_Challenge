### Analysis


 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "../data/Mouse_metadata.csv"
study_results_path = "../data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

df = pd.merge(study_results, mouse_metadata, how = "left")
# Display the data table for preview
df.head()
```

```python
# Checking the number of mice.
df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_ID = df.loc[df.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_info = df.loc[df['Mouse ID'] == 'g989', :]
duplicate_info
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = df[df['Mouse ID'] != 'g989']
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_mean = clean_df['Tumor Volume (mm3)'].groupby(clean_df["Drug Regimen"]).mean()
tumor_median = clean_df['Tumor Volume (mm3)'].groupby(clean_df["Drug Regimen"]).median()
tumor_variance = clean_df['Tumor Volume (mm3)'].groupby(clean_df["Drug Regimen"]).var()
tumor_std = clean_df['Tumor Volume (mm3)'].groupby(clean_df["Drug Regimen"]).std()
tumor_sem = clean_df['Tumor Volume (mm3)'].groupby(clean_df["Drug Regimen"]).sem()
# Assemble the resulting series into a single summary DataFrame.
tumor_summary = pd.DataFrame({
    "Mean Tumor Volume": tumor_mean,
    "Median Tumor Volume": tumor_median,
    "Tumor Volume Variance": tumor_variance,
    "Tumor Volume STD": tumor_std,
    "Tumor Volume SEM": tumor_sem
})
tumor_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary = clean_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean", "median", "var", "std", "sem"])
summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint = clean_df.groupby("Mouse ID").max()["Timepoint"]
greatest_df = pd.DataFrame(greatest_timepoint)
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
volume_df = greatest_df.merge(clean_df, on = ("Mouse ID", "Timepoint"), how = "left")
volume_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volume = df.loc[df["Drug Regimen"]==treatment, "Tumor Volume (mm3)"]
    
    # add subset 
    tumor_list.append(tumor_volume)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq 
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = tumor_volume.loc[(tumor_volume < lower_bound) | (tumor_volume > upper_bound)]
    if outliers.empty:
        print(f"{treatment} has O potential outliers")
    else:
        print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig,ax = plt.subplots()
ax.set_ylabel("Final Tumor Volume (mm3)")
ax.boxplot(tumor_list, labels = treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
# capomulin_df = clean_df.loc[clean_df["Drug Regimen"]=="Capomulin",:]
b742_df = clean_df.loc[clean_df["Mouse ID"]=="b742",:]
x_axis = b742_df["Timepoint"]
y_axis = b742_df["Tumor Volume (mm3)"]
fig, ax = plt.subplots()
plt.title("Capomulin treatment of mouse b742")
plt.plot(x_axis, y_axis, color = "r")
plt.xlabel("Days")
plt.ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]
average_tumor = pd.DataFrame(capomulin_df.groupby(["Mouse ID"])["Weight (g)", "Tumor Volume (mm3)"].mean())

fig, ax = plt.subplots()
plt.scatter(average_tumor["Weight (g)"], average_tumor["Tumor Volume (mm3)"], color = "red" )
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(average_tumor["Weight (g)"],average_tumor["Tumor Volume (mm3)"])
x_values = average_tumor["Weight (g)"]
y_values = average_tumor["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5,1),fontsize=15,color="red")
plt.show()
```
